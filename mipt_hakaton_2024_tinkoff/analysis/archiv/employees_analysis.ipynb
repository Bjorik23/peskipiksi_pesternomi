```python
employees = pd.read_csv('data/employees.csv', sep =';')
employees.head()
```

```python
# @title region vs age

from matplotlib import pyplot as plt
employees.plot(kind='scatter', x='region', y='age', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title exp_days vs edu_degree

from matplotlib import pyplot as plt
employees.plot(kind='scatter', x='exp_days', y='edu_degree', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title age vs exp_days

from matplotlib import pyplot as plt
employees.plot(kind='scatter', x='age', y='exp_days', s=32, alpha=.8)
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title edu_degree

from matplotlib import pyplot as plt
employees['edu_degree'].plot(kind='hist', bins=20, title='edu_degree')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title edu_degree

from matplotlib import pyplot as plt
employees['department_id'].plot(kind='hist', bins=20, title='department_id')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title exp_days

from matplotlib import pyplot as plt
employees['exp_days'].plot(kind='hist', bins=20, title='exp_days')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title sex

from matplotlib import pyplot as plt
import seaborn as sns
employees.groupby('sex').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title sex

from matplotlib import pyplot as plt
import seaborn as sns
employees.groupby('work_online_flg').size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title sex vs region

from matplotlib import pyplot as plt
import seaborn as sns
figsize = (12, 1.2 * len(employees['sex'].unique()))
plt.figure(figsize=figsize)
sns.violinplot(employees, x='region', y='sex', inner='box', palette='Dark2')
sns.despine(top=True, right=True, bottom=True, left=True)
```

```python
# @title age

from matplotlib import pyplot as plt
employees['age'].plot(kind='hist', bins=20, title='age')
plt.gca().spines[['top', 'right',]].set_visible(False)
```

```python
# @title region

from matplotlib import pyplot as plt
employees['region'].plot(kind='hist', bins=20, title='region')
plt.gca().spines[['top', 'right',]].set_visible(False)
```
